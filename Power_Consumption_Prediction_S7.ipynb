```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simulating data for energy consumption based on different factors
np.random.seed(42)
n_samples = 100

# Features: Temperature (in degrees Celsius), Humidity (%), Wall Area (m^2), Roof Area (m^2), Wind Speed (m/s)
temperature = np.random.uniform(0, 40, n_samples)
humidity = np.random.uniform(10, 90, n_samples)
wall_area = np.random.uniform(50, 300, n_samples)
roof_area = np.random.uniform(20, 150, n_samples)
wind_speed = np.random.uniform(0, 20, n_samples)

# Energy consumption (kWh) simulated as a function of these features
energy_consumption = (0.5 * (temperature - 20)**2 + 0.3 * humidity
                      + 0.05 * wall_area + 0.07 * roof_area
                      + 0.1 * wind_speed + np.random.normal(0, 5, n_samples))

# Scatter plot: Temperature vs Energy Consumption (showing U-shaped relationship)
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
sns.scatterplot(x=temperature, y=energy_consumption)
plt.title('Temperature vs Energy Consumption')
plt.xlabel('Temperature (C)')
plt.ylabel('Energy Consumption (kWh)')

# Scatter plot: Wall Area vs Energy Consumption (linear relationship)
plt.subplot(2, 2, 2)
sns.scatterplot(x=wall_area, y=energy_consumption)
plt.title('Wall Area vs Energy Consumption')
plt.xlabel('Wall Area (m)')
plt.ylabel('Energy Consumption (kWh)')

# Scatter plot: Wind Speed vs Energy Consumption
plt.subplot(2, 2, 3)
sns.scatterplot(x=wind_speed, y=energy_consumption)
plt.title('Wind Speed vs Energy Consumption')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Energy Consumption (kWh)')

# 3D plot (temperature, wind speed vs energy consumption) would require more setup, so we'll use a heatmap for now
# Heatmap: Temperature and Humidity (interaction effect on energy consumption)
plt.subplot(2, 2, 4)
data = np.vstack([temperature, humidity]).T
energy_grid = energy_consumption.reshape(n_samples // 10, 10) # Simplifying for heatmap
sns.heatmap(energy_grid, cmap='coolwarm', cbar_kws={'label': 'Energy Consumption (kWh)'})
plt.title('Interaction: Temp & Humidity vs Energy')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming temperature, energy_consumption, and other features are arrays

# Create a DataFrame to work with
df = pd.DataFrame({
    'Temperature': temperature,
    'Humidity': humidity,
    'EnergyConsumption': energy_consumption
})

# Bin the temperature data into 5C intervals
bins = np.arange(0, 45, 5)  # Bins for temperature (0-40C, in steps of 5)
df['TempBin'] = pd.cut(df['Temperature'], bins)

# Calculate the mean energy consumption for each bin
binned_data = df.groupby('TempBin').agg({'EnergyConsumption': 'mean'}).reset_index()

# Plot the binned data
plt.figure(figsize=(8, 5))
sns.barplot(x='TempBin', y='EnergyConsumption', data=binned_data, palette="coolwarm")
plt.title('Average Energy Consumption by Temperature Bin')
plt.xlabel('Temperature (C)')
plt.ylabel('Average Energy Consumption (kWh)')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.show()
```
